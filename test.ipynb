```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from scipy.sparse import hstack


# Cargar los datos en un DataFrame de Pandas
data = pd.read_excel('datos_dengue.xlsx')

# Preparar los datos para el entrenamiento



X = data[['AÃ±o', 'Mes', 'Municipio']] # características
y = data['Numero_casos'] # variable objetivo



# Codificar las variables categóricas
#le_municipio = LabelEncoder()
#['Municipio'] = le_municipio.fit_transform(X['Municipio'])
#ohe_mes = OneHotEncoder()
#mes_ohe = ohe_mes.fit_transform(X['Mes'].values.reshape(-1, 1)).toarray()
#X = pd.concat([X, pd.DataFrame(mes_ohe, columns=['Mes_'+str(i) for i in range(1,13)])], axis=1)
#X = X.drop(['Mes'], axis=1)

#Dividir los datos en conjuntos de entrenamiento y prueba
X_entrenamiento, X_test, y_entrenamiento, y_test = train_test_split(X, y,test_size=25, random_state=42)

# Crear una instancia de la codificación one-hot
#encoder = OneHotEncoder()
#encoder.fit(X_entrenamiento[['Municipio']])
#X_train_codificado = encoder.transform(X_entrenamiento[['Municipio']])
#X_test_codificado = encoder.transform(X_test[['Municipio']])

#print(X_train_codificado)
#print(X_test_codificado)



#X_train_completo = hstack([X_train_codificado, X_entrenamiento[['AÃ±o', 'Mes']]])
#X_test_completo = hstack([X_test_codificado, X_test[['AÃ±o', 'Mes']]])



# Entrenar el modelo
modelo = LinearRegression()
modelo.fit(X_entrenamiento, y_entrenamiento)
```

```python
# Hacer predicciones sobre los datos de prueba
y_pred = modelo.predict(X_test)
```

```python
# Evaluar el modelo en el conjunto de prueba
from sklearn.metrics import r2_score

y_pred = modelo.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f'Error cuadrático medio: {mse}')

r2 = r2_score(y_test, y_pred)

print(r2)
```
